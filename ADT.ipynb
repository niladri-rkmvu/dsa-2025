```python
class Stack:
    def __init__(self):
        self._items = []  # The internal implementation is hidden from users

    def push(self, item):
        self._items.append(item)  # Adds to the top of the stack

    def pop(self):
        if not self.is_empty():
            return self._items.pop()
        raise IndexError("Pop from empty stack")

    def peek(self):
        if not self.is_empty():
            return self._items[-1]
        raise IndexError("Peek from empty stack")

    def is_empty(self):
        return len(self._items) == 0

    def size(self):
        return len(self._items)
```

```python
s = Stack()
s.push(10)
s.push(20)
print(s.peek())  # 20
s.pop()
print(s.size())  # 1
```
