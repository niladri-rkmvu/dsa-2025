# Bubble Sort

```python
class Sorter:
    def __init__(self, arr):
        self.arr = list(arr)

    def print_array(self):
        print(*self.arr)

    def swap(self, i, j):
        self.arr[i], self.arr[j] = self.arr[j], self.arr[i]

    def reset(self, new_arr):
        self.arr = list(new_arr)

    def bubble_sort(self):
        n = len(self.arr)
        for i in range(n - 1):
            flag = False
            for j in range(n - i - 1):
                if self.arr[j] > self.arr[j + 1]:
                    self.swap(j, j + 1)
                    flag = True
            if not flag:
                break

    # def selection_sort(self):
    #     n = len(self.arr)
    #     for i in range(n):
    #         min_idx = i
    #         for j in range(i + 1, n):
    #             if self.arr[j] < self.arr[min_idx]:
    #                 min_idx = j
    #         self.swap(i, min_idx)

    # def insertion_sort(self):
    #     n = len(self.arr)
    #     for i in range(1, n):
    #         key = self.arr[i]
    #         j = i - 1
    #         while j >= 0 and self.arr[j] > key:
    #             self.arr[j + 1] = self.arr[j]
    #             j -= 1
    #         self.arr[j + 1] = key

if __name__ == "__main__":
    arr = [64, 34, 25, 12, 22, 11, 90]
    sorter = Sorter(arr)

    print("Original array:")
    sorter.print_array()

    sorter.bubble_sort()
    print("Bubble sorted:")
    sorter.print_array()

    # reset the array
    sorter.reset(arr)

    # sorter.reset(arr)
    # sorter.selection_sort()
    # print("Selection sorted:")
    # sorter.print_array()

    # sorter.reset(arr)
    # sorter.insertion_sort()
    # print("Insertion sorted:")
    # sorter.print_array()
```
