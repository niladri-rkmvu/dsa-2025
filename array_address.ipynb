# 2D array

```python
# Assign Variables

base_address = 1000
width = 4

# ----------------
#   Address Cord
# ----------------
i = 6
j = 4

# ----------------
# X Coord
# ----------------
x_0 = -3
x_f = 7
X = x_f - x_0 + 1

# ----------------
# Y Coord
# ----------------
y_0 = -5
y_f = 9
Y = y_f - y_0 + 1

print(f"X = {X}, Y = {Y}")
print("---------------")
# ----------------
#       RMO
# ----------------
print(" ")
print("---- RMO ------")

rmo_no_of_preceding_elem = ((i - x_0) * Y) + (j - y_0)

print(f"rmo_no_of_preceding_elem = {rmo_no_of_preceding_elem}")

rmo_address = base_address + rmo_no_of_preceding_elem * width

print(f"rmo_address = {rmo_address}")

print("---------------")

# ----------------
#       CMO
# ----------------
print(" ")
print("---- CMO ------")

cmo_no_of_preceding_elem = ((j - y_0) * X) + (i - x_0)

print(f"cmo_no_of_preceding_elem = {cmo_no_of_preceding_elem}")

cmo_address = base_address + cmo_no_of_preceding_elem * width

print(f"cmo_address = {cmo_address}")

print("---------------")
```

# 3D Array

```python
# Assign Variables

base_address = 1000
width = 4

# ----------------
#   Address Cord
# ----------------
i = 8
j = 4
k = 5

# ----------------
# X Coord
# ----------------
x_0 = 5
x_f = 10
X = x_f - x_0 + 1

# ----------------
# Y Coord
# ----------------
y_0 = 2
y_f = 5
Y = y_f - y_0 + 1

# ----------------
# Z Coord
# ----------------
z_0 = 3
z_f = 6
Z = z_f - z_0 + 1

print(f"X = {X}, Y = {Y}, Z = {Z}")
print("---------------")
# ----------------
#       RMO
# ----------------
print(" ")
print("---- RMO ------")

rmo_no_of_preceding_elem = ((i - x_0) * (Y*Z)) + ((j - y_0)*Z) + (k - z_0)

print(f"rmo_no_of_preceding_elem = {rmo_no_of_preceding_elem}")

rmo_address = base_address + rmo_no_of_preceding_elem * width

print(f"rmo_address = {rmo_address}")

print("---------------")

# ----------------
#       CMO
# ----------------
print(" ")
print("---- CMO ------")

cmo_no_of_preceding_elem = ((k - z_0) * (Y*X)) + ((j - y_0)*X) + (i - x_0)

print(f"cmo_no_of_preceding_elem = {cmo_no_of_preceding_elem}")

cmo_address = base_address + cmo_no_of_preceding_elem * width

print(f"cmo_address = {cmo_address}")

print("---------------")
```
