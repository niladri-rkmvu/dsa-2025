# Heap Sort

```python
def insert(H, n):
    i = n
    temp = H[i]
    while i > 1 and temp > H[i // 2]:
        H[i] = H[i // 2]
        i = i // 2
    H[i] = temp

def delete(H, n):
    root = H[1]
    H[1] = H[n]
    i = 1
    j = 2 * i
    while j < n:
        if j + 1 < n and H[j + 1] > H[j]:
            j += 1
        if H[i] < H[j]:
            H[i], H[j] = H[j], H[i]
            i = j
            j = 2 * i
        else:
            break
    return root

def heapify(H, n, i):
    largest = i
    left = 2 * i
    right = 2 * i + 1

    if left <= n and H[left] > H[largest]:
        largest = left
    if right <= n and H[right] > H[largest]:
        largest = right

    if largest != i:
        H[i], H[largest] = H[largest], H[i]
        heapify(H, n, largest)


def find_violation(H, n):
    for i in range(1, n // 2 + 1):
        left = 2 * i
        right = 2 * i + 1
        if left <= n and H[i] < H[left]:
            return i
        if right <= n and H[i] < H[right]:
            return i
    return None

def test_auto_heapify():
    H = [0, 5, 10, 30, 20, 35, 40, 15]
    n = len(H) - 1

    print("Before auto-heapify:")
    print(H[1:])

    # Apply heapify from bottom-up
    for i in range(n // 2, 0, -1):
        heapify(H, n, i)

    print("After auto-heapify:")
    print(H[1:])

def main():
    H = [0, 10, 20, 30, 25, 5, 40, 35]
    n = len(H) - 1

    for i in range(2, n + 1):
        insert(H, i)

    for i in range(n, 1, -1):
        print(f"Deleted value : {delete(H, i)}")

    print("Sorted array:")
    print(H[1:])

if __name__ == "__main__":
    # main()
    test_auto_heapify()
```
