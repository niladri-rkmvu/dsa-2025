```python
# -------------------------------
# 1.b. Diagonal matrix
# -------------------------------
class Diagonal:
    def __init__(self, n=2):
        # Initialize order and allocate diagonal array
        self.n = n
        self.A = [0] * n

    def set(self, i, j, x):
        # Set element if indices are on the diagonal
        if i == j and 1 <= i <= self.n:
            self.A[i - 1] = x

    def get(self, i, j):
        # Retrieve element if it's diagonal
        if i == j and 1 <= i <= self.n:
            return self.A[i - 1]
        return 0

    def display(self):
        for i in range(self.n):
            for j in range(self.n):
                if i == j:
                    print(self.A[i], end=' ')
                else:
                    print(0, end=' ')
            print()


# Main test block
d = Diagonal(4)

d.set(1, 1, 5)
d.set(2, 2, 8)
d.set(3, 3, 9)
d.set(4, 4, 12)

d.display()
print("M[2,2] =", d.get(2, 2))
```
