```python
# ---------------------------------------------------------------
# Polynomial - create, evaluation, addition
# ---------------------------------------------------------------

from math import pow

class Term:
    def __init__(self, coeff=0, exp=0):
        self.coeff = coeff
        self.exp = exp

class Poly:
    def __init__(self):
        self.terms = []

    def create(self):
        self.n = int(input("Enter the number of non-zero terms : "))
        print("Enter the terms :")
        for i in range(self.n):
            print(f"Term : {i+1}")
            coeff = int(input("Coeff : "))
            exp = int(input("Exp : "))
            self.terms.append(Term(coeff, exp))
            print("----------")

    def display(self):
        for idx, term in enumerate(self.terms):
            if term.exp == 0:
                print(f"{term.coeff}", end='')
            else:
                print(f"{term.coeff}x^{term.exp}", end='')
            if idx != len(self.terms) - 1:
                print("+", end='')
        print()

    def evaluate(self, x):
        sum = 0
        for term in self.terms:
            sum += term.coeff * pow(x, term.exp)
        return sum

# creating a standalone function
def add(p1, p2):
    result = Poly()
    i = j = 0

    while i < len(p1.terms) and j < len(p2.terms):
        if p1.terms[i].exp > p2.terms[j].exp:
            result.terms.append(p1.terms[i])
            i += 1
        elif p2.terms[j].exp > p1.terms[i].exp:
            result.terms.append(p2.terms[j])
            j += 1
        else:
            combined = Term()
            combined.exp = p1.terms[i].exp
            combined.coeff = p1.terms[i].coeff + p2.terms[j].coeff
            result.terms.append(combined)
            i += 1
            j += 1

    result.terms.extend(p1.terms[i:])
    result.terms.extend(p2.terms[j:])
    return result

# Main logic
if __name__ == "__main__":
    p1 = Poly()
    p2 = Poly()

    p1.create()
    p1.display()

    p2.create()
    p2.display()

    # eval_value = 4
    # eval_result = p1.evaluate(eval_value)
    # print(f"Polynomial value for {eval_value} is : {eval_result}")

    p3 = add(p1, p2)
    p3.display()
```
