#General Recursion

```python
def fun(n):
    print(f"called fun(n) => {n}")
    if n > 0:
        fun(n-1)
        print(n)

def main():
    x = 3
    fun(x)

if __name__ == "__main__":
    main()
```

#Head recursion

```python
#***************************************
#    2. Head recursion
#***************************************

def fun(n):
    if n > 0:
        fun(n-1)
        print(n, end=' ')

def main():
    x = 3
    fun(x)

if __name__ == "__main__":
    main()
```

#Tail recursion

```python
#***************************************
#    3. Tail recursion
#***************************************

def fun(n):
    if n > 0:
        print(n, end=' ')
        fun(n-1)

def main():
    x = 3
    fun(x)

if __name__ == "__main__":
    main()
```

#Global variable

```python
#***************************************
#    4. global variable
#***************************************
x = 0  # Global variable
def fun(n):
    global x
    if n > 0:
        x += 1
        return fun(n - 1) + x
    return 0

def main():
    a = 5
    ret_val = fun(a)
    print(ret_val)

    ret_val = fun(a)
    print(ret_val)

if __name__ == "__main__":
    main()
```

#Indirect recursion

```python
#****************************
#  5. Indirect recursion
#*****************************
def functionA(n):
    if n > 0:
        print(f"A: {n}")
        functionB(n - 1)

def functionB(n):
    if n > 0:
        print(f"B: {n}")
        functionA(n // 2)  # Use integer division to mimic C's integer behavior

def main():
    # Start the recursion
    functionA(10)

if __name__ == "__main__":
    main()
```

#Nested recursion

```python
#****************************
#  6. Nested recursion
#*****************************
def fun(n):
    if n > 100:
        return n - 10
    else:
        return fun(fun(n + 11))

def main():
    result = fun(95)
    print("Result:", result)

if __name__ == "__main__":
    main()
```

#Sum of n natural numbers

```python
#****************************
# 7a. Sum of natural numbers
# Method = recursion
# Time = O(n)
# Space = O(n)
#*****************************
def sum_recursive(n):
    if n == 0:
        return 0
    else:
        return sum_recursive(n - 1) + n

def main():
    result = sum_recursive(7)
    print(f"Result = {result}")

if __name__ == "__main__":
    main()
```

```python
#****************************
# 7b. Sum of natural numbers
# Method = Direct formula use
# Time = O(1)
# Space = O(1)
#*****************************
def sum_natural(n):
    return (n * (n + 1)) // 2

result = sum_natural(7)
print(f"Result = {result}")
```

```python
#****************************
#  7c. Sum of natural numbers using for loop = O(n)
#   Time = O(n)
#   Space = O(1)
#*****************************
def sum_iterative(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

def main():
    result = sum_iterative(7)
    print(f"Result = {result}")

if __name__ == "__main__":
    main()
```

#Factorial of a number

```python
#****************************
# 8a. Factorial of a number
#  Method : Recursion
#  Time = O(n)
#  Space = O(n)
#*****************************
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)

def main():
    x = 5
    result = fact(x)
    print(f"{x}! = {result}")

if __name__ == "__main__":
    main()
```

```python
#****************************
#  8b. Factorial of a number
#  Method : using for loop
#  Time = O(n)
#  Space = O(1)
#*****************************
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

def main():
    num = 5
    result = factorial(num)
    print(f"{num}! = {result}")

if __name__ == "__main__":
    main()
```

#Exponent

```python
#****************************
#  9a. Exponent
#  Method   : Recursion
#  Time     : O(n)
#  Space    : O(n)
#*****************************
def power_recursive(m, n):
    if n == 0:
        return 1
    return m * power_recursive(m, n - 1)

# Example usage
m, n = 2, 5
print(f"{m}^{n} = {power_recursive(m, n)}")
```

```python
#****************************
#  9a. Exponent
#  Method   : Using Loop
#  Time     : O(n)
#  Space    : O(1)
#*****************************
def power_iterative(m, n):
    result = 1
    for _ in range(n):
        result *= m
    return result

# Example usage
m, n = 2, 5
print(f"{m}^{n} = {power_iterative(m, n)}")
```

```python
#****************************
#  9a. Exponent
#  Method   : Exponent Squaring
#  Time     : O(log n)
#  Space    : O(log n)
#*****************************

def power_fast(m, n):
    if n == 0:
        return 1
    if n % 2 == 0:
        return power_fast(m * m, n // 2)
    else:
        return m * power_fast(m * m, (n - 1) // 2)

# Example usage
m, n = 2, 10
print(f"{m}^{n} = {power_fast(m, n)}")
```

#Taylor Series



$$
e^x = \sum_{n=0}^{\infty} \frac{x^n}{n!} = 1 + \frac{x}{1!} + \frac{x^2}{2!} + \frac{x^3}{3!} + \cdots
$$

```python
#****************************
#  10.a. Taylor series | Recursion
#*****************************

# Global variables to track power and factorial
p = 1.0
f = 1.0

def e(x, n):
    global p, f
    if n == 0:
        return 1
    r = e(x, n - 1)
    p *= x
    f *= n
    return r + (p / f)

def main():
    global p, f
    x = 2  # Power of e
    n = 6  # Number of terms

    # Reset globals before calling the function
    p = 1.0
    f = 1.0

    result = e(x, n)
    print(f"Function = e^{x} | No. of terms = {n} | Result = {result:.5f}")

if __name__ == "__main__":
    main()
```

```python
#****************************
#  10.b. Taylor series | Iterative
#*****************************
def e(x, n):
    s = 1.0
    num = 1.0  # numerator (x^i)
    den = 1.0  # denominator (i!)

    for i in range(1, n + 1):
        num *= x
        den *= i
        s += num / den

    return s

# Example usage
print(f"{e(2, 10):.6f}")
```

```python
import math

def e(x, n):
    s = 1.0
    num = 1.0
    den = 1.0
    for i in range(1, n + 1):
        num *= x
        den *= i
        s += num / den
    return s

x = 2
n = 10

approx = e(x, n)
actual = math.exp(x)
error = abs(actual - approx)

print(f"e^{x} ≈ {approx:.6f} using {n} terms")
print(f"Built-in math.exp({x}) = {actual:.6f}")
print(f"Absolute error = {error:.6e}")
```

```python
import matplotlib.pyplot as plt
import math

def taylor_e(x, n_terms):
    s = 1.0
    num = 1.0
    den = 1.0
    approximations = [1.0]

    for i in range(1, n_terms + 1):
        num *= x
        den *= i
        s += num / den
        approximations.append(s)

    return approximations

x = 2
n_terms = 20
approx_values = taylor_e(x, n_terms)
actual_value = math.exp(x)

plt.plot(range(n_terms + 1), approx_values, label='Taylor Approximation', marker='o')
plt.axhline(y=actual_value, color='r', linestyle='--', label=f'Actual e^{x} = {actual_value:.5f}')
plt.title(f'Convergence of Taylor Series for e^{x}')
plt.xlabel('Number of Terms')
plt.ylabel('Approximate Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#****************************
#  10.c. Taylor series | Horner's rule
#  Multiplication time = O(n)
#*****************************
s = 1.0  # acts like the static variable in C++
def e(x, n):
    global s
    if n == 0:
        return s
    s = 1 + (x * s) / n
    return e(x, n - 1)

# Example usage
if __name__ == "__main__":
    s = 1.0  # reset global static-like accumulator
    print(f"{e(2, 10):.6f}")
```

#Fibonacci series

```python
# 11.a. Fibonacci series
# Method: Loop (iterative)

def fibonacci(n):
    t0, t1 = 0, 1

    if n <= 1:
        return n

    for _ in range(2, n + 1):
        s = t0 + t1
        t0 = t1
        t1 = s

    return s

# Main block
number = 6
result = fibonacci(number)
print(f"The {number}th Fibonacci number is {result}")
```

```python
# 11.b. Fibonacci series
# Method: Recursion

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Main block
number = 6
result = fibonacci(number)
print(f"The {number}th Fibonacci number is {result}")
```

```python
# 11.c. Fibonacci series
# Method: Recursion using memoization

F = [-1] * 10  # Initialize memoization array with -1

def mfib(n):
    if n <= 1:
        F[n] = n
        return n
    else:
        if F[n - 2] == -1:
            F[n - 2] = mfib(n - 2)
        if F[n - 1] == -1:
            F[n - 1] = mfib(n - 1)
        return F[n - 2] + F[n - 1]

# Main block
result = mfib(5)
print(f"Result of mfib(5) = {result}")
```

#Combination
$$
nCr = \binom{n}{r} = \frac{n!}{r!(n - r)!}
$$

```python
# 12.a. nCr
# Method: Formula based using factorial (recursive)

def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)

def C(n, r):
    if r < 0 or r > n:
        return -1
    return fact(n) // (fact(r) * fact(n - r))

# Main block
print(f"C(4, 2) = {C(4, 2)}")
```

```python
# 12.b. nCr
# Method: Using Pascal triangle (recursive)

def C(n, r):
    if r < 0 or r > n:
        return -1

    # base cases
    if r == 0 or n == r:
        return 1
    else:
        return C(n - 1, r - 1) + C(n - 1, r)

# Main block
print(f"C(4, 2) = {C(4, 2)}")
```

# Tower of Hanoi

```python
def TOH(source, destination, auxiliary,n):
    if n == 1:
        print(f"Move only disk from {source} to {destination}")
        return
    else:
        TOH(source, auxiliary, destination, n-1)
        print(f"Move top disk {n} from {source} to {destination}")
        TOH(auxiliary, destination, source, n-1)

# Example usage:
num_disks = 3
TOH('A', 'B', 'C', num_disks)
```
