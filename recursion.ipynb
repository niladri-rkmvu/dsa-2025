```python
def fun(n):
    print(f"called fun(n) => {n}")
    if n > 0:
        fun(n-1)
        print(n)

def main():
    x = 3
    fun(x)

if __name__ == "__main__":
    main()
```

```python
#***************************************
#    2. Head recursion
#***************************************

def fun(n):
    if n > 0:
        fun(n-1)
        print(n, end=' ')

def main():
    x = 3
    fun(x)

if __name__ == "__main__":
    main()
```

```python
#***************************************
#    3. Tail recursion
#***************************************

def fun(n):
    if n > 0:
        print(n, end=' ')
        fun(n-1)

def main():
    x = 3
    fun(x)

if __name__ == "__main__":
    main()
```

```python
#***************************************
#    4. global variable
#***************************************
x = 0  # Global variable like in the original C code
def fun(n):
    global x
    if n > 0:
        x += 1
        return fun(n - 1) + x
    return 0

def main():
    a = 5
    ret_val = fun(a)
    print(ret_val)

    ret_val = fun(a)
    print(ret_val)

if __name__ == "__main__":
    main()
```

```python
#****************************
#  5. Indirect recursion
#*****************************
def functionA(n):
    if n > 0:
        print(f"A: {n}")
        functionB(n - 1)

def functionB(n):
    if n > 0:
        print(f"B: {n}")
        functionA(n // 2)  # Use integer division to mimic C's integer behavior

def main():
    # Start the recursion
    functionA(10)

if __name__ == "__main__":
    main()
```

```python
#****************************
#  6. Nested recursion
#*****************************
def fun(n):
    if n > 100:
        return n - 10
    else:
        return fun(fun(n + 11))

def main():
    result = fun(95)
    print("Result:", result)

if __name__ == "__main__":
    main()
```

```python
#****************************
# 7a. Sum of natural numbers
# Method = recursion
# Time = O(n)
# Space = O(n)
#*****************************
def sum_recursive(n):
    if n == 0:
        return 0
    else:
        return sum_recursive(n - 1) + n

def main():
    result = sum_recursive(7)
    print(f"Result = {result}")

if __name__ == "__main__":
    main()
```

```python
#****************************
# 7b. Sum of natural numbers
# Method = Direct formula use
# Time = O(1)
# Space = O(1)
#*****************************
def sum_natural(n):
    return (n * (n + 1)) // 2

result = sum_natural(7)
print(f"Result = {result}")
```

```python
#****************************
#  7c. Sum of natural numbers using for loop = O(n)
#   Time = O(n)
#   Space = O(1)
#*****************************
def sum_iterative(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

def main():
    result = sum_iterative(7)
    print(f"Result = {result}")

if __name__ == "__main__":
    main()
```

```python
#****************************
# 8a. Factorial of a number
#  Method : Recursion
#  Time = O(n)
#  Space = O(n)
#*****************************
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)

def main():
    x = 5
    result = fact(x)
    print(f"{x}! = {result}")

if __name__ == "__main__":
    main()
```

```python
#****************************
#  8b. Factorial of a number
#  Method : using for loop
#  Time = O(n)
#  Space = O(1)
#*****************************
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

def main():
    num = 5
    result = factorial(num)
    print(f"{num}! = {result}")

if __name__ == "__main__":
    main()
```

```python
#****************************
#  9a. Exponent
#  Method   : Recursion
#  Time     : O(n)
#  Space    : O(n)
#*****************************
def power_recursive(m, n):
    if n == 0:
        return 1
    return m * power_recursive(m, n - 1)

# Example usage
m, n = 2, 5
print(f"{m}^{n} = {power_recursive(m, n)}")
```

```python
#****************************
#  9a. Exponent
#  Method   : Using Loop
#  Time     : O(n)
#  Space    : O(1)
#*****************************
def power_iterative(m, n):
    result = 1
    for _ in range(n):
        result *= m
    return result

# Example usage
m, n = 2, 5
print(f"{m}^{n} = {power_iterative(m, n)}")
```

```python
#****************************
#  9a. Exponent
#  Method   : Exponent Squaring
#  Time     : O(log n)
#  Space    : O(log n)
#*****************************

def power_fast(m, n):
    if n == 0:
        return 1
    if n % 2 == 0:
        return power_fast(m * m, n // 2)
    else:
        return m * power_fast(m * m, (n - 1) // 2)

# Example usage
m, n = 2, 10
print(f"{m}^{n} = {power_fast(m, n)}")
```
