#Factorial

```python
def fact(n):
  if n <= 1:
    return 1
  else:
    # reduced value of n, also just immediate next step
    return n * fact(n-1)

# Example usage
number = 5
result = fact(number)
print(f"The factorial of {number} is {result}")
```

#Fibonacci
0,1,1,2,3,5,8

```python
def fibonacci(n):
  if n <= 1:
    return n
  else:
    # reduced value of n, also just immediate next step
    return fibonacci(n-1) + fibonacci(n-2)

# Example usage
number = 6
result = fibonacci(number)
print(f"The {number}th Fibonacci number is {result}")
```
