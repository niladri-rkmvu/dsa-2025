```python
# ----------------------------------
# 1.a Create a sparse matrix | Python
# ----------------------------------

class Element:
    def __init__(self, i: int, j: int, x: int):
        self.i = i  # represents row
        self.j = j  # represents col
        self.x = x  # represents val

class Sparse:
    def __init__(self):
        self.m = 0  # dimension of matrix = row
        self.n = 0  # dimension of matrix = col
        self.num = 0  # number of non-zero elements
        self.ele = []  # list to store Element instances

    def create(self):
        self.m, self.n = map(int, input("Enter Dimensions of the matrix (m n): ").split())
        print()

        self.num = int(input("Number of non-zero elements: "))
        print()

        print("Enter non-zero elements (row, col, val): ")
        for i in range(self.num):
            i_, j_, x_ = map(int, input(f"Element[{i}] [row col val]: ").split())
            self.ele.append(Element(i_, j_, x_))

    def display(self):
        print("\nDisplaying matrix:")
        k = 0
        for i in range(self.m):
            for j in range(self.n):
                # check if current position matches the next element's position
                if k < self.num and i == self.ele[k].i and j == self.ele[k].j:
                    print(self.ele[k].x, end=" ")
                    k += 1
                else:
                    print(0, end=" ")
            print()

# -------------------------
# Execution
# -------------------------
if __name__ == "__main__":
    s = Sparse()
    s.create()
    s.display()
```

```python
# ----------------------------------
# 2. Create a sparse matrix | Python
# Include add function
# Ensure you enter the input from 0 index - for both row and col
# ----------------------------------

class Element:
    def __init__(self, i, j, x):
        self.i = i  # row index
        self.j = j  # column index
        self.x = x  # value

class Sparse:
    def __init__(self):
        self.m = 0       # matrix rows
        self.n = 0       # matrix columns
        self.num = 0     # number of non-zero elements
        self.ele = []    # list of Element objects

    def create(self):
        self.m, self.n = map(int, input("Enter Dimensions of matrix (m n): ").split())
        print()

        self.num = int(input("Enter the number of non-zero elements (should be less than 50% of mÃ—n): "))
        print()

        print("Enter non-zero elements (row col val) in sorted order or follow row-major order:")
        for idx in range(self.num):
            print(f"Element[{idx}].[row col val] = ", end="")
            row, col, val = map(int, input().split())
            self.ele.append(Element(row, col, val))

    def display(self):
        print("\nDisplaying matrix:")
        k = 0
        for i in range(self.m):
            for j in range(self.n):
                if k < self.num and self.ele[k].i == i and self.ele[k].j == j:
                    print(self.ele[k].x, end=" ")
                    k += 1
                else:
                    print(0, end=" ")
            print()

def add_sparse(s1, s2):
    if s1.m != s2.m or s1.n != s2.n:
        return None

    result = Sparse()
    result.m = s1.m
    result.n = s1.n

    i = j = 0
    while i < s1.num and j < s2.num:
        e1 = s1.ele[i]
        e2 = s2.ele[j]

        if (e1.i < e2.i) or (e1.i == e2.i and e1.j < e2.j):
            result.ele.append(Element(e1.i, e1.j, e1.x))
            i += 1
        elif (e1.i > e2.i) or (e1.i == e2.i and e1.j > e2.j):
            result.ele.append(Element(e2.i, e2.j, e2.x))
            j += 1
        else:
            result.ele.append(Element(e1.i, e1.j, e1.x + e2.x))
            i += 1
            j += 1

    while i < s1.num:
        e = s1.ele[i]
        result.ele.append(Element(e.i, e.j, e.x))
        i += 1
    while j < s2.num:
        e = s2.ele[j]
        result.ele.append(Element(e.i, e.j, e.x))
        j += 1

    result.num = len(result.ele)
    return result

# -------------------------
# Execution
# -------------------------
if __name__ == "__main__":
    print("Matrix 1:")
    s1 = Sparse()
    s1.create()
    s1.display()

    print("Matrix 2:")
    s2 = Sparse()
    s2.create()
    s2.display()

    result = add_sparse(s1, s2)
    if result:
        print("\nResultant Matrix After Addition:")
        result.display()
    else:
        print("Cannot add matrices: dimension mismatch.")
```
